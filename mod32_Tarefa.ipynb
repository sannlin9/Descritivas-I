# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math
import seaborn as sns

import matplotlib.ticker as mtick
from scipy.stats import norm, t
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
fig, ax = plt.subplots(figsize=(10,2))

mu = 15
variance = .1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.cdf(x, mu, sigma)
ax.plot(x, y)

std_lim = 0 # 95% CI
low = 15.2
high = 16

plt.fill_between(x, y, where=(low < x) & (x < high), alpha=.5)
# plt.text(low, 0, low, ha='center')
# plt.text(high, 0, high, ha='center')

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))

prob_area = norm.cdf(high, mu, sigma) - norm.cdf(low, mu, sigma)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
percentil = norm.ppf(0.95, mu, variance)

print("O percentil de ordem 95% para o peso dos pacotes  {:.2f} kg".format(percentil))
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

mu = 0
variance = 1

distribuicao_Y1 = norm(loc=mu, scale=variance)
distribuicao_Y2 = t(df=5)

probabilidades_Y1 = distribuicao_Y1.cdf(Y)
probabilidades_Y2 = distribuicao_Y2.cdf(Y)

print("Probabilidades para Y1:")
for i in range(len(Y)):
    print("P(Y1 < {:.2f}) = {:.4f}".format(Y[i], probabilidades_Y1[i]))

print("\nProbabilidades para Y2:")
for i in range(len(Y)):
    print("P(Y2 < {:.2f}) = {:.4f}".format(Y[i], probabilidades_Y2[i]))
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
plt.plot(Y, probabilidades_Y1, label='Y1 ~ Normal(0,1)')
plt.fill_between(Y, probabilidades_Y1, alpha=0.2)

plt.plot(Y, probabilidades_Y2, label='Y2 ~ Distancia T(5)')
plt.fill_between(Y, probabilidades_Y2, alpha=0.2)

plt.legend()
plt.show()
```

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

percentis_Y1 = distribuicao_Y1.ppf(Y)
percentis_Y2 = distribuicao_Y2.ppf(Y)

print("Percentis para Y1:")
for i in range(len(Y)):
    print("{:.2f} : {:.4f}".format(Y[i], percentis_Y1[i]))

print("\nPercentis para Y2:")
for i in range(len(Y)):
    print("{:.2f} : {:.4f}".format(Y[i], percentis_Y2[i]))
```
