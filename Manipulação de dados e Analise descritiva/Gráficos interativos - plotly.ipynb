```python
%%time
ativos = ['ITUB3.SA', 'PETR4.SA', 'VALE3.SA']
dt_ini = '2021-01-01'
dt_fim = '2023-04-01'

yf = yfin.download(ativos, dt_ini,dt_fim)

yf.head()
```

```python
yf.to_csv('ativos_bovespa.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
yf_high = yf['High']
yf_high = yf_high.stack().reset_index()
yf_high.columns = ['data', 'ativo', 'high']
yf_high
```

```python
yf_low = yf['Low']
yf_low = yf_low.stack().reset_index()
yf_low.columns = ['data', 'ativo', 'low']
yf_low
```

```python
yf_ampli = yf_high
```

```python
yf_ampli['low'] = yf_low['low']
yf_ampli['amplitude'] = yf_ampli['high'] - yf_ampli['low']
yf_ampli
```

```python
# Visual Python: Visualization > Plotly
fig = px.line(yf_ampli, x='data', y='amplitude', color='ativo', template='plotly_white')
fig.show()
```

Nos dois ultimos anos a Vale (VALE3) apresentou a maior amplitudes de ações dentro de um mesmo dia.
