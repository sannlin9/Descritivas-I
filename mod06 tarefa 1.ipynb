```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

O arquivo PDF com o dicionario dos dados está disponivel em: <a href="https://static.lms.ebaconline.com.br/content-files/9b580b8a-8ebc-4796-936d-999a59f33d17/original.pdf"> Estrutura_SINASC_para_CD.pdf <a>

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=sinasc)
```

```python
sinasc['GESTACAO'].value_counts().sort_index().plot.bar()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5',y = 'CONSULTAS', data=sinasc)
```
